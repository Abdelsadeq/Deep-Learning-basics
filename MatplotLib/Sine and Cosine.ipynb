```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0, 3*np.pi, 0.1)
y= np.sin(x)
z = np.cos(x)

plt.title("Sine and Cosine Functions")
plt.legend(['sine','Cosine'])
plt.xlabel("X axis")
plt.ylabel("y axis")
plt.plot(x,y)
plt.plot(x,z)
```
